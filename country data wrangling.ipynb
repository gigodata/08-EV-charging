```python
import pandas as pd
```

```python
annual = pd.read_csv('./Resources/average annual wage.csv')
location = pd.read_csv('./Resources/country location.csv')
minimum = pd.read_csv('./Resources/real minimum wage.csv')
```

```python
location.rename(columns={'name' : 'Country',
                         'latitude' : 'Latitude',
                         'longitude' : 'Longitude'}, inplace=True)
location = location[['Country', 'Latitude', 'Longitude']]
location.head()
```

```python
annual = annual[['Country', 'Time', 'Value']]
annual.rename(columns={'Value' : 'Average Annual Income'}, inplace=True)
annual.head()
```

```python
minimum = minimum[['Country', 'Time', 'Value']]
minimum.rename(columns={'Value' : 'Real Hourly Minimum Wage'}, inplace=True)
minimum.head()
```

```python
newdf = pd.merge(minimum, annual, on=['Country', 'Time'])
newdf.head()
```

```python
fullDf = pd.merge(newdf, location, on='Country')
fullDf.head()
```

```python
fullDf.to_csv('../08-EV-charging/Resources/cleanedData.csv')
```
