```python
import pandas as pd
```

```python
annual = pd.read_csv('./Resources/average annual wage.csv')
location = pd.read_csv('./Resources/country location.csv')
minimum = pd.read_csv('./Resources/real minimum wage.csv')
coffee = pd.read_csv('./Resources/Coffee_domestic_consumption.csv')
```

```python
coffee = coffee[['Country', '2001/02', '2002/03', '2003/04', '2004/05', '2005/06', '2006/07', '2007/08', '2008/09', '2009/10', '2010/11', '2011/12', '2012/13', '2013/14',
                '2014/15', '2015/16', '2016/17', '2017/18', '2018/19', '2019/20']]
coffee.head()
```

```python
coffee.rename(inplace=True, columns={'2001/02' : '2001',
                                     '2002/03' : '2002',
                                     '2003/04' : '2003', 
                                     '2004/05' : '2004', 
                                     '2005/06' : '2005', 
                                     '2006/07' : '2006', 
                                     '2007/08' : '2007', 
                                     '2008/09' : '2008', 
                                     '2009/10' : '2009', 
                                     '2010/11' : '2010', 
                                     '2011/12' : '2011', 
                                     '2012/13' : '2012', 
                                     '2013/14' : '2013', 
                                     '2014/15' : '2014', 
                                     '2015/16' : '2015', 
                                     '2016/17' : '2016', 
                                     '2017/18' : '2017', 
                                     '2018/19' : '2018', 
                                     '2019/20' : '2019'})
coffee.head()
```

```python
coffee = coffee.melt(id_vars='Country',
            var_name='Time',
            value_name='Coffee Consumption', ignore_index=True)
coffee.head()
```

```python
coffee['Time'] = pd.to_numeric(coffee['Time'])
coffee['Time'].dtype
```

```python
location.rename(columns={'name' : 'Country',
                         'latitude' : 'Latitude',
                         'longitude' : 'Longitude'}, inplace=True)
location = location[['Country', 'Latitude', 'Longitude']]
location.head()
```

```python
annual = annual[['Country', 'Time', 'Value']]
annual.rename(columns={'Value' : 'Average Annual Income'}, inplace=True)
annual.head()
```

```python
minimum = minimum[['Country', 'Time', 'Value']]
minimum.rename(columns={'Value' : 'Real Hourly Minimum Wage'}, inplace=True)
minimum.head()
```

```python
newdf = pd.merge(minimum, annual, on=['Country', 'Time'])
newdf.head()
```

```python
fullDf = pd.merge(newdf, location, on='Country')
fullDf.head()
```

```python
fullDf = pd.merge(fullDf, coffee, on=['Country', 'Time'])
fullDf.head()
```

```python
fullDf.dropna(inplace=True)
fullDf.shape
```

```python
fullDf.to_csv('../08-EV-charging/Resources/cleanedData.csv')
```
